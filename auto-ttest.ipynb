```python
from scipy.stats import ttest_ind,norm,f
import numpy as np
def ftest(s1,s2):
    '''F检验样本总体方差是否相等'''
    #print("Null Hypothesis:var(s1)=var(s2)，α=0.05")
    F = np.var(s1)/np.var(s2)
    v1 = len(s1) - 1
    v2 = len(s2) - 1
    p_val = 1 - 2*abs(0.5-f.cdf(F,v1,v2))
    #print(p_val)
    if p_val < 0.05:
        #print("Reject the Null Hypothesis.")
        equal_var=False
    else:
        #print("Accept the Null Hypothesis.")
        equal_var=True
    return equal_var
        
def ttest_ind_fun(s1,s2):
    '''t检验独立样本所代表的两个总体均值是否存在差异'''
    equal_var = ftest(s1,s2)
    #print("Null Hypothesis:mean(s1)=mean(s2)，α=0.05")
    ttest,pval = ttest_ind(s1,s2,equal_var=equal_var)
    #if pval < 0.05:
        #print("Reject the Null Hypothesis.")
    #else:
        #print("Accept the Null Hypothesis.")
    return ttest,pval

def desc_ttest(df,var,group,v1,v2):
    g1 = np.array(df[var][df[group] == v1 ])
    g2 = np.array(df[var][df[group] == v2 ])
    t =  ttest_ind_fun(g1,g2)[0]
    p =  ttest_ind_fun(g1,g2)[1]
    dfn = df.groupby(group)
    a = pd.DataFrame(dfn[var].mean())
    b = pd.DataFrame(dfn[var].std())
    c = pd.DataFrame(dfn[var].count())
    d = pd.DataFrame({'t 值':[t],'P 值':[p]})
    d.index = [v1]
    e = pd.concat([c,a,b,d],axis =1)
    e = e.reset_index()
    e.sort_values(by='t 值',ascending = True,inplace = True)
    e.columns = ['组别','样本量','均值','标准差','t 值','P 值']
    e = e.replace(np.nan,' ')
    e.style.hide_index().format('{:.3f}',subset=['均值','标准差','t 值','P 值'])
    print("\n下为变量{}成组t检验结果:".format(var))
    return print(e)

#desc_ttest(df,'Age','GROUP',1,2)
#desc_ttest(df,'BMI','GROUP',1,2)

list0 = df.columns.to_list()
list1 = ['Number', 'Sex','GROUP']
list2 = [x for x in list0 if x not in list1]

for i in range(len(list2)):
    desc_ttest(df,list2[i],'GROUP',1,2)
for i in range(len(list2)):
    desc_ttest(df,list2[i],'Sex','Male','Female')    
```
